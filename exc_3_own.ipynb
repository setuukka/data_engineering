```python
#3.Use company database from Exercise 2. Use SQLite and query the daily total amount of sales from table orders. Make a line plot of daily sales using matplotlib.pyplot
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("company_database.db")
```

```python
#ORDERS: order_id,customer_id,order_date,amount

query = '''

SELECT
order_date as date,
sum(amount) as total
FROM
orders
GROUP BY order_date



'''
query_results = pd.read_sql(query, conn)
query_results.head()
```

```python
df = pd.read_sql_query(query, conn)
df['date'] = pd.to_datetime(df['date'])
plt.figure(figsize = (12,6))
plt.bar(df['date'], df['total'])
plt.title("Daily total sales")
plt.show()
```
