2: The zip package contains of seven .csv files. Download the package, extract the files and carry out the following:

a) Create SQL database company_database.db where each csv file appears as a table. Use SQLite

b) Query project id, employee id and hours worked from employee_projects and employee id's and salaries from 
employees_realistic. Join them on employee id.

c) Calculate the salary costs per project, assuming that the employers salary is annual and number of working hours is 1900 per year. 

d) Join this with the budget in projects table

e) Calculate how large fraction of the project budget the salary costs are.


You can use VS code SQL or Python SQLite in the task 2 b - e. 
Example of the output of the query in 2e

```python
import pandas as pd
import sqlite3
```

```python
#a) Create SQL database company_database.db where each csv file appears as a table. Use SQLite
#Read each csv to pandas dataframe
df_customers = pd.read_csv("company data/customers.csv")
df_departments = pd.read_csv("company data/departments.csv")
df_employee_projects = pd.read_csv("company data/employee_projects.csv")
df_employees_realistic = pd.read_csv("company data/employees_realistic.csv")
df_order_items = pd.read_csv("company data/order_items.csv")
df_orders = pd.read_csv("company data/orders.csv")
df_projects = pd.read_csv("company data/projects.csv")
```

```python
#creating database
conn = sqlite3.connect("company_database.db")
#create table from each dataframe
df_customers.to_sql('customers', conn, if_exists = 'replace', index = False)
df_departments.to_sql('departments', conn, if_exists = 'replace', index = False)
df_employee_projects.to_sql('employee_projects', conn, if_exists = 'replace', index = False)
df_employees_realistic.to_sql('employees_realistic', conn, if_exists = 'replace', index = False)
df_order_items.to_sql('order_items', conn, if_exists = 'replace', index = False)
df_orders.to_sql('orders', conn, if_exists = 'replace', index = False)
df_projects.to_sql('projects', conn, if_exists = 'replace', index = False)
```

```python
#b) Query project id, employee id and hours worked from employee_projects and employee id's and salaries from 
#employees_realistic. Join them on employee id.

query = '''
SELECT ep.employee_id, hours_worked, er.salary
FROM
employee_projects as ep
JOIN
employees_realistic AS er
ON ep.employee_id = er.employee_id

'''
query_results = pd.read_sql(query, conn)
query_results.head()
```

```python
#c) Calculate the salary costs per project, assuming that the employers salary is annual and number of working hours is 1900 per year. 


#d) Join this with the budget in projects table

#e) Calculate how large fraction of the project budget the salary costs are.


query = '''
SELECT 
p.project_id, 
p.budget, 
round(sum(ep.hours_worked * er.salary / 1900),2) as project_salary_cost,
round(sum(ep.hours_worked * er.salary / 1900) / p.budget * 100,2) as percentage

FROM
projects as p
JOIN
employee_projects as ep
ON p.project_id = ep.project_id
JOIN
employees_realistic as er
ON ep.employee_id = er.employee_id


GROUP BY p.project_id

'''
query_results = pd.read_sql(query, conn)
query_results.head()
```
