```python
#Query data from BigQuery: Select query
#Download data from BigQuery
import pandas_gbq

projectid = 'automated-shape-317108'
#Assume that enviroment variable GOOGLE_APPLICATION_CREDENTIALS 
#points the path to Google Cloud credentials key file

#Query
sql = '''SELECT * FROM `bigquery-public-data.ml_datasets.penguins`'''

#Read the data
df =  pandas_gbq.read_gbq(sql,projectid)
```

```python
#Some data manipulation
df = df[df['island'] == "Dream"]
df
```

```python
# Upload this table to BigQuery using API.

### Save results to BigQuery
dataset ='test_data_2025'
table_name = 'Dream_island_penguins'
destination = dataset + '.'+ table_name
#Export data to BigQuery

df.to_gbq(destination, project_id = projectid, if_exists = 'replace') #Method 1
```

```python
#Do the same thing by Action query
#Data is not moved from BigQuery at any point

from google.cloud import bigquery
#Use GOOGLE_APPLICATION_CREDENTIALS
client = bigquery.Client()

#Set table_id to the ID of the destination table.
projectid = 'automated-shape-317108'
dataset ='test_data_2025_US' #Dataset to save date in US
#In case of action query, source data and target data must be in the same location!
table_name = 'Dream_island_penguins_aq' #Table to save data
destination = projectid + '.' + dataset + '.'+ table_name #Destination to save data


job_config = bigquery.QueryJobConfig(destination=destination)
job_config.write_disposition = "WRITE_TRUNCATE" #Overwrite if exist


sql = '''
        SELECT 
        * 
        FROM `bigquery-public-data.ml_datasets.penguins`
        WHERE island = "Dream"
        '''

# Start the query, passing in the extra configuration.
query_job = client.query(sql, job_config=job_config) # Make an API request.
query_job.result() # Wait for the job to complete.
print("Query results loaded to the table {}".format(destination))
```
