```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import glob
import os
```

```python
#Define connection to database
conn = sqlite3.connect("company_database.db")
cursor = conn.cursor()
```

```python
#Reading all csv files in folder to pandas dataframes and store them into dictionary
dfs = {}
for item in os.listdir():
    if item.endswith('csv'):
        filename = f"df_{item.split('.')[0]}"
        #print(filename)
        curr_df = pd.read_csv(item)
        dfs[filename] = curr_df
```

```python
#Reading all dataframes to sqlite database tables
for key, df in dfs.items():
    df.to_sql(name = key, if_exists = 'replace', con = conn)
```

```python
#Confirming that tables exist
for row in cursor.execute("SELECT name FROM sqlite_master WHERE type='table';"):
    print(row)
```
